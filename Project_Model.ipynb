```python
df = pd.read_csv('XYZCorp_LendingData.txt',delimiter = "\t",low_memory=False)
df.head(3)
```

```python
print(df.shape)
print(df.isnull().sum())
```

## Dropping columns with more than 50% missing values and unwanted variables

```python
 

unwanted_cols = ['annual_inc_joint','application_type','collections_12_mths_ex_med',
                 'collections_12_mths_ex_med','desc','dti_joint','earliest_cr_line','emp_title',
                 'funded_amnt_inv','id','member_id','mths_since_last_delinq',
                 'mths_since_last_major_derog','mths_since_last_record','open_acc',
                 'out_prncp_inv','policy_code','pub_rec','pymnt_plan','recoveries','sub_grade',
                 'title','total_pymnt_inv','total_rec_int','total_rec_late_fee','total_rec_prncp',
                 'verification_status_joint','zip_code','open_acc_6m','open_il_6m','open_il_12m',
                 'open_il_24m','mths_since_rcnt_il','total_bal_il','il_util','open_rv_12m',
                 'open_rv_24m','max_bal_bc','all_util','total_rev_hi_lim','inq_fi','total_cu_tl',
                 'inq_last_12m','acc_now_delinq']


df_copy = df.drop(unwanted_cols,axis=1)
```

```python
df_copy.shape
```

```python
df_copy.isnull().sum()
```

```python
#imputing missing values with the mean
col_mean = ['revol_util','tot_coll_amt','tot_cur_bal']
for name in col_mean:
    df_copy[name].fillna(df_copy[name].mean(),inplace=True)
print(df_copy.isnull().sum())
```

```python
#imputing missing values with the mode
unwant = ['last_pymnt_d','emp_length','last_credit_pull_d','next_pymnt_d']
for name in unwant:
    df_copy[name].fillna(df_copy[name].mode()[0], inplace=True)
print(df_copy.isnull().sum())
```

```python
df_copy.dtypes
```

```python
colname = ['loan_amnt','term','grade','emp_length','home_ownership',
           'verification_status','purpose','addr_state','initial_list_status',
           'last_pymnt_d','next_pymnt_d','last_credit_pull_d']
```

```python
df_copy['issue_d']= pd.to_datetime(df_copy.issue_d)
df_copy.dtypes
```

```python
df_copy.set_index('issue_d', inplace=True)
df_copy.head(3)
```

```python
df_copy.sort_index(ascending=True,inplace=True)
```

```python
df_copy.head()
```

```python
from sklearn import preprocessing

le={}

for x in colname:
    le[x]=preprocessing.LabelEncoder()
    
    
for x in colname:
    df_copy[x]=le[x].fit_transform(df_copy.__getattr__(x))
```

```python
df_copy.head(5)
```

# Train Test Split

```python
split_index = df_copy.index
```

```python
type(split_index)
```

```python
split_index[2]
```

```python
x = pd.to_datetime('2015-05-31')
```

```python
split_index[2] < x
```

```python
X_train = df_copy.iloc[df_copy.index < x, :]
X_train.shape
```

```python
X_test = df_copy.iloc[df_copy.index > x, :]
X_test.shape
```

```python
Y_train=X_train.values[:,-1]
Y_train.shape
```

```python
X_train=X_train.values[:,:-1]
X_train.shape
```

```python
Y_test=X_test.values[:,-1]
print(Y_test.shape)

X_test=X_test.values[:,:-1]
X_test.shape
```

# =================================

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
#print(X)
```

# Model Building Decision Tree

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train,Y_train)

Y_pred =model.predict(X_test)
#print(list(zip(Y_test,Y_pred)))
```

```python
cfm = confusion_matrix(Y_test,Y_pred)
print(cfm)

print('Classification report:')
print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test,Y_pred)
print("accuracy of the model:",acc)
```

```python
from sklearn import metrics
fpr, tpr, threshold = metrics.roc_curve(Y_test,Y_pred)
auc = metrics.auc(fpr, tpr)
print('Area Under the Curve:', auc)
```

```python
# Model Building Random Forest
```
